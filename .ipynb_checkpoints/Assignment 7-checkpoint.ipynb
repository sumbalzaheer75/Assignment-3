### Employees details with OOP

```python
class Employee():
    def __init__(self, name, number, address, salary, post, year_of_joining, education, year_of_experience):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.year_of_experience = year_of_experience
    def describe_employee(self):
        print("\n")
        print(f"Name : {self.name}")
        print(f"Number : {self.number}")
        print(f"Address : {self.address}")
        print(f"Salary : {self.salary}")
        print(f"Post : {self.post}")
        print(f"Year of joining : {self.year_of_joining}")
        print(f"Education : {self.education}")
        print(f"Year of experience : {self.year_of_experience}")
        print("\n")
    def update_year_of_experience(self, new_experience):
        self.year_of_experience = new_experience
        print(f"Updated year of experience to : {new_experience}")
    def increase_salary(self, increment):
        self.salary += increment
        print(f"New salary after increment : {self.salary}")
employee101 = Employee("Salar", 234, "Karachi", 120000, "Assistant Manager", 2021, 'BSCS', 2)
print(employee101.name)
print(employee101.address)
print(employee101.salary)
print(employee101.post)
employee101.describe_employee()
employee101.update_year_of_experience(3)
employee101.increase_salary(5000)
employees = [
    Employee("Sara", 200, "Lhr", 60000, "HR", 2023, "BS", 1),
    Employee("Haris", 201, 'Isb', 30000, 'Assistant', 2024, 'Graduation', 0),
    Employee("Hina", 202, 'Khi', 90000, 'Supervisor', 2022, 'BSIT', 2),
    Employee("Salman", 203, 'Pindi', 150000, 'Manager', 2020, 'BBA', 4),
    Employee("Rehan", 204, 'Haripur', 200000, 'Branch Manager', 2019, 'MBA', 5),
    Employee("Noman", 205, 'Hunza', 80000, 'AM', 2021, 'BE', 3),
    Employee("Maham", 206, 'Multan', 50000, 'Admin', 2023, 'BSCS', 1),
    Employee("Fiza", 207, 'Bakhar', 750000, 'Operation Manager', 2021, 'BBA', 3),
    Employee("Hashir", 208, 'Gawadar', 60000, 'Sales', 2022, 'BSc', 2),
    Employee("Daniyal", 209, 'Quetta', 30000, 'Internee', 2024, 'BSDS', 0)
]
employee_details_list = [employee.describe_employee() for employee in employees]
employees[4].update_year_of_experience(3)
employees[5].increase_salary(7000)
employees[2].update_year_of_experience(1)
employees[3].increase_salary(5000)
```
